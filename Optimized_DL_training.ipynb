```python
!python -m torch.distributed.launch --nproc_per_node=8 train_amp.py -a resnet50 --b 256 --workers 4 --epochs 5 --lr=0.25 --opt-level O3 ./datasets/imagenet
```

FP16 training with FP32 batchnorm:

Keeping the batchnorms in FP32 improves stability and allows Pytorch to use cudnn batchnorms, which significantly increases speed in Resnet50.

```python
!python -m torch.distributed.launch --nproc_per_node=8 train_amp.py -a resnet50 --b 256 --workers 4 --epochs 5 --lr=0.25 --opt-level O3 --keep-batchnorm-fp32 True ./datasets/imagenet
```

--opt-level O1 (Official Mixed Precision recipe, recommended for typical use)

GEMM = FP16

risky layers = FP32

dynamic loss scaling


```python
!python -m torch.distributed.launch --nproc_per_node=8 train_amp.py -a resnet50 --b 256 --workers 4 --epochs 5 --lr=0.25 --opt-level O1 ./datasets/imagenet
```

Use static loss_scaling

```python
!python -m torch.distributed.launch --nproc_per_node=8 train_amp.py -a resnet50 --b 256 --workers 4 --epochs 5 --lr=0.25 --opt-level O1 --loss-scale 128.0 ./datasets/imagenet
```
